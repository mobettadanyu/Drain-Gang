```python
from threading import Thread
import requests

THREAD_COUNT = 6

def callback():
    try:
        while True:
            r = requests.get('https://docs.google.com/forms/d/e/1FAIpQLSdAf0GBS0HwskIot2YAR1vS5FQ0eYhjKqO48CWAtMoq_0gHlw/formResponse?usp=pp_url&entry.508027384=Hey&submit=SUBMIT')
            print(r)
    except KeyboardInterrupt:
        return

if __name__ == '__main__':
    threads = []
    for i in range(THREAD_COUNT):
        t = Thread(target=callback)
        threads.append(t)
        t.start()

    for t in threads:
        t.join()
```
